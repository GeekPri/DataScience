```python
# Question1: Find if there are any differece in salary when the hsc_board is Central or Others .
#1 condtion vale(salary) 2 group vale(central, others)

from scipy.stats import ttest_ind
dataset = dataset.dropna()
centralGroup = dataset[dataset['hsc_b'] == 'Central']['salary']
othersGroup = dataset[dataset['hsc_b'] == 'Others']['salary']
t_test1 = ttest_ind(centralGroup, othersGroup)
#Ttest_indResult(statistic=-0.30570032095155825, pvalue=0.7601313863865756)
# pvalue> 5%. Not similar 

#
#
# Question2: Find if there are any differece in performance for a commerce student in  mba test and science student in mba_test
#1 condtion vale(mba_test) 2 group vale(commerce, science)
CommerceGroup = dataset[dataset['hsc_s'] == 'Commerce']['mba_p']
ScienceGroup = dataset[dataset['hsc_s'] == 'Science']['mba_p']
t_test2 = ttest_ind(CommerceGroup, ScienceGroup)
#Ttest_indResult(statistic=-0.7331285580404581, pvalue=0.46432995253854314)
#pvalue = 46% and not <5%. Similarity is NOT present. This mean students from Commerce group and Science group have NOT equally scored in mba test 
 
```
